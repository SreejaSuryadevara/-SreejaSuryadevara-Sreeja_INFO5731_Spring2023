```python
# You code here 
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('avocado.csv')
data.head()

```

```python
total_vol=data[data["AveragePrice"]>1.00]
Top_region =total_vol[total_vol["region"]=="Albany"].sort_values(by='AveragePrice',ascending=False)
typee=total_vol[total_vol["type"]=="conventional"].sort_values(by='AveragePrice',ascending=False)
print(Top_region)
print(typee)
```

```python
k=[]
k.append(Top_region ['AveragePrice'].max())
k.append(typee['AveragePrice'].max())
l=["region","type"]
plt.bar(l,k)
plt.title("The Max avocados sold are")
plt.ylabel('Amount')
```

```python
# You code here (Please add comments in the code):
from bs4 import BeautifulSoup
import requests

url = 'https://scholar.google.com/scholar?hl=en&as_sdt=0%2C44&q=avocado&btnG='
response=requests.get(url)
soup=BeautifulSoup(response.content,'lxml')

for item in soup.select('[data-lid]'):
	try:
		print('----------------------------------------')
		print(item)
		
	except Exception as e:
		#raise e
		print('')
for item in soup.select('[data-lid]'): 
  try: 
    print('----------------------------------------') 
    #print(item) 
    print(item.select('h3')[0].get_text()) 
    print(item.select('a')[0]['href']) 
    print(item.select('.gs_rs')[0].get_text()) 
    print('----------------------------------------') 
  except Exception as e:
     print('')

```
